```python
# Dependencies and Setup

import pandas as pd
import numpy as np
```

```python
# Importing data files

census_df = pd.read_csv("resources/census_us_county_output.csv")
education_df = pd.read_csv("resources/ers_usda_education.csv")
county_df = pd.read_csv("resources/county_fips.csv")
```

## Cleaning County_fips.csv

```python
county_df
```

```python
# Reseting index
county_df = county_df.reset_index(drop=True)
county_df
```

```python
county_df = county_df[county_df["County_Name"].str.contains("County")==True]
county_df
```

```python
# column_names = list(df.columns)
# column_names[4:8] = ['five', 'six', 'seven', 'eight']
# df.columns = column_names


column_names = list(county_df.columns)
column_names[0:4] = ['fips_code', 'county', 'state_abbr', 'state']
county_df.columns = column_names

county_df
```

```python
county_df.dropna()
```

```python
# Reseting index
county_df = county_df.reset_index(drop=True)
county_df
```

```python
county_df
```

## Cleaning census_us_county_output.csv

```python
census_df
```

```python
# Reseting index
census_df = census_df.reset_index(drop=True)
```

```python
# Splitting Name column into county name and state

census_df[['County_x','State']] = census_df.Name.str.split(",",expand=True,)
census_df
```

```python
census_df = census_df.drop_duplicates()
census_df
```

```python
# Remove column name 
census_df = census_df.drop(['Name', 'County'], axis = 1)
census_df
```

```python
# column_names = list(df.columns)
# column_names[0:9] = ['five', 'six', 'seven', 'eight']
# df.columns = column_names

column_names = list(census_df.columns)
column_names[0:9] = ['population', 'median_age', 'household_income', 'per_capita_income', 'poverty_count',
                     'poverty_rate', 'unemployment_rate', 'county', 'state']
census_df.columns = column_names
```

```python
census_df
```

```python
# Rename column
# df=df.rename(columns = {'two':'new_name'})


census_df = census_df.rename(columns = {'County_x': 'County_Name'})
census_df
```

```python
census_df = census_df[census_df["county"].str.contains("County")==True]
census_df
```

```python
census_df["county"] = census_df["county"].str.slice(0, -6)
```

```python
census_df = census_df.fillna(0)
census_df
```

```python
census_df = census_df.sort_values(by = ['state', 'county'], ascending = [True, True])
census_df
```

```python
census_df = census_df.replace(' ','', regex=True)
```

```python
census_df.info()
```

## Cleaning ers_usda_education.csv

```python
education_df
```

```python
# to see null rows
# education_df.info()
```

```python
# Reset index
education_df = education_df.reset_index(drop=True)
```

```python
education_df
```

```python
# Rename column
# df=df.rename(columns = {'two':'new_name'})


education_df = education_df.rename(columns = {'FIPS Code': 'FIPS_Code', 'Area name':'County'})
education_df
```

```python
education_df = education_df[education_df["County"].str.contains("County")==True]
```

```python
# Remove extra characters from 'County' name
#education_df["County"] = education_df.County.str.replace('\s.*','')
education_df["County"] = education_df["County"].str.slice(0, -6)
```

```python
# Remove all columns between column index 3 to 31

education_df = education_df.drop(education_df.iloc[:, 3:31], axis = 1)
education_df
```

```python
# column_names = list(df.columns)
# column_names[4:8] = ['five', 'six', 'seven', 'eight']
# df.columns = column_names

column_names = list(education_df.columns)
column_names[0:19] = ['fips_code', 'state_abbr', 'county', 'below_hs_diploma_2000', 'hs_diploma_2000', 'college_or_associate_2000', 'bachelors_or_higher_2000',
                    'percent_below_hs_diploma_2000', 'percent_hs_diploma_2000', 'percent_college_or_associate_2000', 'percent_bachelors_or_higher_2000',
                    'below_hs_diploma_2019', 'hs_diploma_2019', 'college_or_associate_2019', 'bachelors_or_higher_2019',
                    'percent_below_hs_diploma_2019', 'percent_hs_diploma_2019', 'percent_college_or_associate_2019', 'percent_bachelors_or_higher_2019']
education_df.columns = column_names
```

```python
education_df
```

```python
education_df = education_df.drop_duplicates()
education_df
```

```python
education_df = education_df.replace(',','', regex=True)
```

```python
education_df = education_df.fillna(0)
education_df
```

```python
education_df.columns
```

```python
# Merging county and education data

merge_df = pd.merge(county_df, education_df, how='inner', on=['fips_code'])
merge_df
```

```python
# Dropping the duplicate column

merge_df = merge_df.drop(merge_df.iloc[:, 4:6], axis = 1)
merge_df
```

```python
# Renaming the columns

merge_df = merge_df.rename(columns = {'county_x': 'county', 'state_abbr_x':'state_abbr'})
merge_df
```

```python
merge_df.dropna()
```

```python
# Creating new dataframe for fips code, county name and state

fips_df = merge_df.loc[:, ['fips_code', 'county', 'state_abbr', 'state']]
fips_df
```

```python
# Save as a csv
# Note to avoid any issues later, use encoding="utf-8"
fips_df.to_csv("resources/cleaned_county_fips.csv", encoding="utf-8", index=False)
```

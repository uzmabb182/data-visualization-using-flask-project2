```python
# Dependencies and Setup

import pandas as pd
import numpy as np
```

```python
# Importing data files

census_df = pd.read_csv("Resources/census_us_county_output.csv")
education_df = pd.read_csv("Resources/ers_usda_education.csv")
```

```python
census_df
```

```python
# Reseting index
census_df = census_df.reset_index(drop=True)
```

```python
# Remove column name 'A'
census_df = census_df.drop(['County'], axis = 1)
census_df
```

```python
census_df[['County','State']] = census_df.Name.str.split(",",expand=True,)
census_df
```

```python
census_df = census_df.drop_duplicates()
census_df
```

```python
# Save as a csv
# Note to avoid any issues later, use encoding="utf-8"
census_df.to_csv("Resources/cleaned_census_data.csv", encoding="utf-8", index=False)
```

```python
# Reseting index
education_df = education_df.reset_index(drop=True)
```

```python
education_df = education_df.iloc[2:]
education_df
```

```python
# to see null rows
# education_df.info()
```

```python
# Reset index
education_df = education_df.reset_index(drop=True)
```

```python
education_df
```

```python
# Rename column
# df=df.rename(columns = {'two':'new_name'})


education_df = education_df.rename(columns = {'Area name':'County'})
education_df
```

```python
# Remove all columns between column index 3 to 31

education_df = education_df.drop(education_df.iloc[:, 3:31], axis = 1)
education_df
```

```python
# education_df.rename(columns = {'test':'TEST', 'odi':'ODI',
#                               't20':'T20'}, inplace = True)

# column_names = list(df.columns)
# column_names[4:8] = ['five', 'six', 'seven', 'eight']
# df.columns = column_names

column_names = list(education_df.columns)
column_names[3:19] = ['Below_HS_Diploma_2000', 'HS_Diploma_2000', 'College_or_Associate_2000', 'Bachelor_or_Higher_2000',
                    'Percent_Below_HS_Diploma_2000', 'Percent_HS_Diploma_2000', 'Percent_College_or_Associate_2000', 'Percent_Bachelor_or_Higher_2000',
                    'Below_HS_Diploma_2019', 'HS_Diploma_2019', 'College_or_Associate_2019', 'Bachelor_or_Higher_2019',
                    'Percent_Below_HS_Diploma_2019', 'Percent_HS_Diploma_2019', 'Percent_College_or_Associate_2019', 'Percent_Bachelor_or_Higher_2019',]
education_df.columns = column_names
```

```python
education_df
```

```python
education_df = education_df.drop_duplicates()
education_df
```

```python
# Save as a csv
# Note to avoid any issues later, use encoding="utf-8"
education_df.to_csv("Resources/cleaned_education_data.csv", encoding="utf-8", index=False)
```
